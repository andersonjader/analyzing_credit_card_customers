```python
def teste_tukey(dados,features, col_classe):
    for f in features:
        compara_grupo = MultiComparison(df[f], df[col_classe])
        teste = compara_grupo.tukeyhsd()
        print(20*"=")
        print(f)
        print(teste)
        
```

```python
features = ['MolLogP', 'MolWt', 'NumHAcceptors', 'NumHDonors', 'NumRotatableBonds',
       'RingCount', 'TPSA']
```
